```python
points_allowed_df = df[['season_year', 'opp_team', 'week', 'DK points']]
```

```python
def get_last_week(row):
    year = row['season_year']
    team = row['opp_team']
    week = row['week'] - 1
    if week == 0:
        return None
    else:
        val = points_allowed_df[(points_allowed_df['season_year'] == year) & (points_allowed_df['opp_team'] == team) & (points_allowed_df['week'] == week)]
        if len(val['DK points'].values) == 0:
            return None
        else:
            return val['DK points'].values[0]
```
